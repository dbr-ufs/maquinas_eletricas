Conforme visto em aula, existe uma potência dita ativa que "realiza trabalho" e uma potência reativa que "em média" não realiza trabalho. Na prática o que existe é a potência instantânea, em alguns momentos a energia é armazenada e em outro pode vir a ser descarregada no sistema. 

Para exemplificar vamos definir uma impedância composta por uma resistência e uma capacitância:

```octave
R = 120 % Ohm
C = 250E-6 %F
f = 60 % Hz
Xc = 1/(2*pi*f*C) % Ohm
Z = 120 - Xc*i
```

Alimentando essa impedância com uma fonte alternada:

```octave
V = 120 + 0*i % V
I = V / Z % A
abs(I)
angle(I)
```

Sabemos que potência é o produto interno entre a tensão e a corrente:

```octave
P = V * I % W
S = abs(P) % VA
Q = imag(P) % VAr
```

Para exemplificar, pode-se plotar o vetor Potência:

```octave
plot([0 P])
```

## Valor Eficaz

Observa-se acima que a potência não é constante e, para simplificar o cálculo da potência, definimos valor eficaz com o equivalente a um sinal contínuo, ou seja:

$I_{eficaz} = \frac{I_{max}}{\sqrt 2}$, idem pra tensão.

A potência é, portanto, o produto dos modulos e a soma dos ângulos (relembrar produto entre dois vetores na forma polar), resultado em:

$p = V_{max} . I_{max} . cos (w.t + \theta)$

Esquecendo o tempo, para saber o valor de potência ativa em um sinal puramente senoidal, pode-se simplesmente, para cada fase, calcular:

$P = V_{eficaz} . I_{eficaz} . cos (\theta)$
